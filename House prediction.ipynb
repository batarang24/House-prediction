```python

def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
is_float(2)
```

```python
df2[~df2['total_sqft'].apply(lambda x:is_float(x))].head()
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None   
```

```python
df2['sqft']=df2['total_sqft'].apply(lambda x:convert_sqft_to_num(x))
df2.head()
```

```python
df2.isnull().sum()
df2=df2.dropna()
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3=df2
df3=df3.drop(['total_sqft'],axis="columns")
df3.head()
```

```python
df3['price_per_sqft']=df3['price']*100000/df3['sqft']
df3.head()
```

```python
df3['bhk']=df3.bhk.apply(lambda x:int(x))
df3
```

```python
df3[df3.sqft/df3.bhk<300]
```

```python
df3.shape
```

```python
df4=df3[~(df3.sqft/df3.bhk<300)]
df4.shape
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df5 = remove_pps_outliers(df4)
df5.shape
```

```python
dummies=pd.get_dummies(df5.location)
dummies
```

```python
df6=pd.concat([df5,dummies],axis="columns")
df6
```

```python
df7=df6.drop(['location'],axis="columns")
```

```python
X=df7.drop('price',axis="columns")
```

```python
y=df7.price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```
